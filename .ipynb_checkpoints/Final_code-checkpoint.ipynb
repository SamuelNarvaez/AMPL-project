```python
from music21 import * #import the music21 library
import xml.etree.ElementTree as ET
import os
import matplotlib.pyplot as plt
```

```python
with open ("motifs.csv") as f:
    lines = f.readlines()
```

```python
motifs = [line.strip() for line in lines]
motifs
```

```python
motifs = [motif[1:-1].split(", ") for motif in motifs]
motifs
```

```python
"""
def add_cento(cento):
    StreamMot = stream.Stream()
    for c in motif:
        StreamMot.append(note.Note(c))
    return StreamMot
"""
```

```python
all_motifs = []

for motif in motifs:
    StreamMot = stream.Stream()
    for c in motif:
        StreamMot.append(note.Note(c))
    all_motifs.append(StreamMot)
```

```python
for idx, i in enumerate(all_motifs):
    print(f"Motif number: {idx}")
    for n in i.notes:
        print(n)
```

```python
music = converter.parse("xmlfile.xml")
```

```python
n_occ = []

fps = music.recurse().notes

for motif in all_motifs:
    print(motif)
    pfp = search.noteNameSearch(fps, motif)
    n_occ.append(len(pfp))

n_occ
```

```python
# Extend motif list :TODO
alter = ["", "#", "b"]
ext_motifs = [i for i in motifs]

for motif in motifs:
    for idx, _note in enumerate(motif):
        for alteration in alter:
            new_motif = motif
            _note+=alteration
            new_motif[idx] = _note
        ext_motifs.append(new_motif)

ext_motifs


```
